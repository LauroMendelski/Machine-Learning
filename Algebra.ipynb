```python


# 1. Definir os Vetores
v1 = [1, 2]
v2 = [5, 3]

# Escalares menores para visualizao
a = 2
b = 1

# Calcular o Resultante (Sua lgica corrigida)
v_resultante = [
    (a * v1[0]) + (b * v2[0]), 
    (a * v1[1]) + (b * v2[1])
]

# 2. Configurar o Grfico
plt.figure(figsize=(8, 8)) # Tamanho quadrado para no distorcer
plt.grid(True, linestyle='--', alpha=0.6) # Grade para ver as coordenadas
plt.axhline(0, color='black', linewidth=1) # Eixo X
plt.axvline(0, color='black', linewidth=1) # Eixo Y

# Funo auxiliar para no repetir cdigo (DRY - Don't Repeat Yourself)
def plotar_vetor(vetor, cor, nome):
    # plt.arrow(x_origem, y_origem, dx, dy, ...)
    plt.arrow(0, 0, vetor[0], vetor[1], 
              head_width=0.3, head_length=0.3, fc=cor, ec=cor, label=nome, 
              length_includes_head=True)

# 3. Plotar as setas
plotar_vetor(v1, 'blue', 'V1 original')
plotar_vetor(v2, 'red', 'V2 original')
plotar_vetor(v_resultante, 'green', 'Resultante (Combinao)')

# 4. Ajustar Limites (Importante: seno o grfico corta as setas)
plt.xlim(0, 10) # Ajuste conforme o tamanho do maior vetor
plt.ylim(0, 10)

plt.legend()
plt.title(f"Combinao Linear: {a}*v1 + {b}*v2")
plt.show()
```
